```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-201-resources/master/ESS_practice_data/ESSdata_Thinkful.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df_2012=df[df['year']==6]
df_2014=df[df['year']==7]
```

```python
df_2012['ppltrst'].hist()
```

```python
df_2014['ppltrst'].hist()
```

```python
print(df_2012['ppltrst'].mean())
print(df_2014['ppltrst'].mean())
```

```python
#since theyare relatively normal i will proceed with parametric t-testing testing
import scipy.stats as stats
```

```python
tval,pval=stats.ttest_rel(df_2012['ppltrst'],df_2014['ppltrst'],nan_policy='omit')
```

```python
print(tval)
print(pval)
```

```python
# from combination of the graph visuals and the t value and more importantly, the p value, we can conclude that people did not
# become more mistrusting in 2014
```

```python
df_2012=df_2012.dropna()
df_2014=df_2014.dropna()
```

```python
df_2012['happy'].hist()
```

```python
df_2014['happy'].hist()
```

```python
print(df_2012['happy'].mean())
print(df_2014['happy'].mean())
```

```python
tval,pval=stats.wilcoxon(df_2012['happy'],df_2014['happy'][:-5])
```

```python
print(tval)
print(pval)
```

```python
# again as seen visually and by testing, there is very little diference if any, in happiness levels
```
